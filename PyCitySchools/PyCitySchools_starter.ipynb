```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_both_passed=0
Total_Schools=(school_data_complete["school_name"].nunique())
TotaL_Students=(school_data_complete["Student ID"].nunique())
Total_Budget=(school_data_complete["budget"].sum())
Average_Maths_Score=(school_data_complete["maths_score"].mean())
Average_Reading_Score=(school_data_complete["reading_score"].mean())
Pass_Maths=(len(school_data_complete[school_data_complete["maths_score"] >= 50])/TotaL_Students)*100
Pass_Reading=(len(school_data_complete[school_data_complete["reading_score"] >= 50])/TotaL_Students)*100

for x, y in school_data_complete.iterrows():
   if y['maths_score'] >= 50 and y['reading_score'] >= 50:
       total_both_passed += 1

Pass_Overall = (total_both_passed / len(school_data_complete)) * 100


district_summary_df = pd.DataFrame({"Total Schools": [Total_Schools],
"Total Students": [TotaL_Students],"Total Budget": [Total_Budget],"Average Math Score": [Average_Maths_Score],"Average Reading Score": [Average_Reading_Score],"% Passing Math": [Pass_Maths],"% Passing Reading": [Pass_Reading],"% Overall Passing Rate": [Pass_Overall]})
district_summary_df.head()
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_name=school_data_complete.groupby("school_name")
school_type=school_name["type"].first()
total_students=school_name["Student ID"].nunique() 
total_budget=school_name["budget"].first()
per_student_budget=total_budget/total_students
avg_maths_score=school_name["maths_score"].mean()
avg_reading_score=school_name["reading_score"].mean()


percent_math_passing = ((school_data_complete[school_data_complete["maths_score"] >=50].groupby("school_name")["maths_score"].count()) / total_students) * 100
percent_reading_passing=((school_data_complete[school_data_complete["reading_score"] >=50].groupby("school_name")["reading_score"].count()) / total_students) * 100
Pass_Overall_2=(percent_math_passing+percent_reading_passing)/2

school_summary=pd.DataFrame({"School Type": school_type, "Total Students": total_students, "Total School Budget": Total_Budget, "Per Student Budget":per_student_budget,"Average Math Score": avg_maths_score, "Average Reading Score": avg_reading_score,"% Passing Math":percent_math_passing,"% Reading Passing": percent_reading_passing, "% Overall Passing Rate": Pass_Overall_2})

school_summary.head()
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
Btm_performing=school_summary.sort_values(by="% Overall Passing Rate",ascending=False)
Btm_performing.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
summary_2=school_summary.copy()
bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

summary_2["score_spending"] = pd.cut(summary_2['Per Student Budget'], bins=bins, labels=labels)
b_avg_maths_score = summary_2.groupby("score_spending")['Average Math Score'].mean()
b_avg_reading_score = summary_2.groupby("score_spending")['Average Reading Score'].mean()
b_percent_maths_passing=summary_2.groupby("score_spending")['% Passing Math'].mean()
b_percent_reading_passing=summary_2.groupby("score_spending")['% Reading Passing'].mean()
b_Pass_Overall=(b_percent_maths_passing+b_percent_reading_passing)/2

score_spending=pd.DataFrame({"Average Math Score":b_avg_maths_score,"Average Reading Score":b_avg_reading_score,"% Passing Math":b_percent_maths_passing,"% Passing Reading":b_percent_reading_passing,"% Overall Passing Rate":b_Pass_Overall})
score_spending.head()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
labels_2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

summary_2["School Size"] = pd.cut(summary_2['Total Students'], bins=size_bins, labels=labels_2)

c_avg_maths_score = summary_2.groupby("School Size")['Average Math Score'].mean()
c_avg_reading_score = summary_2.groupby("School Size")['Average Reading Score'].mean()
c_percent_maths_passing=summary_2.groupby("School Size")['% Passing Math'].mean()
c_percent_reading_passing=summary_2.groupby("School Size")['% Reading Passing'].mean()
c_Pass_Overall=(c_percent_maths_passing+c_percent_reading_passing)/2

score_spending=pd.DataFrame({"Average Math Score":c_avg_maths_score,"Average Reading Score":c_avg_reading_score,"% Passing Math":c_percent_maths_passing,"% Passing Reading":c_percent_reading_passing,"% Overall Passing Rate":c_Pass_Overall})
score_spending.head()
```
